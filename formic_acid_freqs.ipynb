```python
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
x = []
y = []
z = []

f = open ('/Users/user/Desktop/Comp_Chem/formic_acid/formic_acid_freqs.csv' , 'r')
freqs = csv.reader(f, delimiter = ',')
for row in freqs:
    x.append(row[0])
    y.append(row[1])
    z.append(row[2])

for i in range(0,len(y)):
    y[i]=int(y[i])

for i in range(0,len(z)):
    z[i]=int(z[i])

plt.stem(x,y)
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
